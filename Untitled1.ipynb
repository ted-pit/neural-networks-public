```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5, 5, 3000)
y1=[]
y2=[]
for i in range(len(x)):
  if x[i] >0:
    y1.append(x[i])
  else:
    y1.append(x[i]*-2)

f = plt.figure()
f.set_figwidth(8)
f.set_figheight(3)
plt.plot(x, y1)
plt.show()

plt.show()
```

```python
r = plt.figure()
r.set_figwidth(8)
r.set_figheight(3)
plt.plot(x,y2)
plt.show()

plt.show()
```
