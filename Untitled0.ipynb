```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import dmol
soldata = pd.read_csv(
    "https://github.com/whitead/dmol-book/raw/master/data/curated-solubility-dataset.csv"
)
```

```python
features_start_at = list(soldata.columns).index("MolWt")
feature_names = soldata.columns[ features_start_at:]
labels = soldata.Solubility.values
f = soldata[feature_names]
fmean = np.mean(f, axis=0)
fstd = np.std(f, axis=0)
features = ((f - fmean) / fstd)

data = tf.data.Dataset.from_tensor_slices((features.values, labels))
trainSize = int(0.85 * len(features))
testData = data.skip(trainSize).batch(16)
trainData = data.take(trainSize).batch(16)

hiddenLayer = tf.keras.layers.Dense(32, activation="tanh")  ## 32 neurons in the layer
outputLayer = tf.keras.layers.Dense(1)
model = tf.keras.Sequential()
model.add(hiddenLayer)
model.add(outputLayer)

model.compile(optimizer="SGD", loss="mean_squared_error") ## Finalise model

model.fit(trainData, epochs=60)
```

```python
yhat = np.squeeze(model.predict(testData))
test_y = labels[:trainSize]
plt.plot(test_y, yhat, ".")
plt.plot(test_y, test_y, "-")
plt.xlabel("Measured Solubility $y$")
plt.ylabel("Predicted Solubility $\hat{y}$")
plt.text(
    min(test_y) + 1,
    max(test_y) - 2,
    f"correlation = {np.corrcoef(test_y, yhat)[0,1]:.3f}",
)
plt.text(
    min(test_y) + 1,
    max(test_y) - 3,
    f"loss = {np.sqrt(np.mean((test_y - yhat)**2)):.3f}",
)
plt.show()
```
