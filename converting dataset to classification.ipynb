```python
#### HOMO

data=pd.read_csv('gdrive/MyDrive/dataset/dataset_cleansed.csv')
```

```python
print(min(data.HOMO))
print(max(data.HOMO))
```

```python
homo=[]
for i in range(len(data.HOMO)):
  homo.append(data.HOMO[i])
homo.sort(reverse=True)
low_cutoff = int(len(homo) * 0.8)
high_cutoff = int(len(homo) * 0.2)
low_homo = homo[low_cutoff]
high_homo = homo[high_cutoff]
```

```python
count =0
for i in range(len(data.HOMO)):
  if data.HOMO[i] > homo[high_cutoff]:
    count = count+1
print(count/len(homo)) #what proportion of the dataset lies within the defined threshold
```

```python
label=[]
for i in range(len(data.HOMO)):
  if data.HOMO[i] > high_homo: 
    label.append(1)
  else:
    label.append(0)

```

```python
data['label']= label
data2 = data.drop('HOMO', axis=1)
```

```python
data2.to_csv('gdrive/MyDrive/dataset/dataset_classification_highHOMO.csv')
```

```python
shuffled_dataset = data2.sample(frac=1).reset_index(drop=True)
shuffled_dataset = shuffled_dataset.drop('Unnamed: 0', axis=1)

train_proportion = 0.8
train_length = (int(train_proportion*len(shuffled_dataset)))
test_length = len(shuffled_dataset) - train_length

train_dataset = shuffled_dataset[:train_length]
test_dataset = shuffled_dataset[train_length:]
```

```python
train_dataset.to_csv("gdrive/MyDrive/dataset/dataset_classification_highHOMO_train.csv")
test_dataset.to_csv("gdrive/MyDrive/dataset/dataset_classification_highHOMO_test.csv")
```
