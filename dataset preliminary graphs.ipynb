```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from google.colab import drive 
drive.mount('/content/gdrive')
dataset = pd.read_csv("gdrive/MyDrive/dataset/CSD_EES_DB.csv")  ##maybe implement standardising the data as a transformer?
```

```python
f, (ax1, ax2) = plt.subplots(1,2)
f.set_figwidth(10)
f.set_figheight(4)
ax1.scatter(dataset.HOMO, dataset.LUMO, 2.8)
ax1.set(xlabel="HOMO /eV",ylabel="LUMO /eV")

ax2.hist(dataset.HOMO, bins=60, alpha=0.7, color='Green', label='HOMO')
ax2.hist(dataset.LUMO, bins=60, alpha=0.7, color='Red', label='LUMO')
ax2.legend()
ax2.set(xlabel="E /eV",ylabel="count")




plt.show()
```
